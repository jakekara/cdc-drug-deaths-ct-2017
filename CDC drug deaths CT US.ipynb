```python
def increase(st, m1="December",m2="December",y1=2016,y2=2017,just_numbers=True, value="pr_int"):
    ret = df[(df["State"] == st) & (df["Indicator"] == "Number of Drug Overdose Deaths")
             & ((df["Year"] == y1) | (df["Year"] == y2))
             * ((df["Month"] == m1) | (df["Month"] == m2))].sort_values("Year")
    
    if not just_numbers:
        return ret
    
    ret = pd.pivot_table(ret[["State","Year","pr_int","dv_int"]], columns="Year", values=value,index="State")
    
    ret["incr"] = (ret[y2] - ret[y1]) 

    ret["pct"] = ret["incr"] / ret[y1]
    
    return ret



increase("US")
```

```python
increase("CT")
```

```python
ne_states = ["CT","RI","ME","NH","MA","VT"]
df[df["State"].apply(lambda x: x in ne_states + ["US"])]["State Name"].unique()
```

```python
def states_increase(value="pr_int", state_set=states):
    frames = []
    
    for st in state_set:
        frames.append(increase(st, value=value))
        
    ret = pd.concat(frames)
    
    return ret

states_increase(value="dv_int", state_set=ne_states + ["US"]).sort_values(by="incr", ascending=False)
```

```python
states_increase(value="pr_int", state_set=ne_states + ["US"]).sort_values(by="pct", ascending=False)
```

```python
states_increase(value="dv_int").sort_values(by="pct", ascending=False)
```

```python
months = ["January","February","March","April","May","June",
         "July","August","September","October","November","December"]
print len(months)
```

```python
def month_no(mon):
    return months.index(mon) + 1

month_no("January")
month_no("December")
```

```python
df["month_no"] = df["Month"].apply(month_no)
```

```python
#July 1 2017 est
pop = {
    "US": 325719178,
    "CT": 3588184
}
```

```python
def in_order(st, value=["pr_int"], adjust=False):
    ret = df[(df["State"] == st) & (df["Indicator"] == "Number of Drug Overdose Deaths")].sort_values(by=["Year","month_no"])
    ret["mno"] = ret["Year"].astype(str) + "-" + ret["month_no"].astype(str)
    
    
    ret = ret.set_index("mno")
    
    ret = ret[value] 
    
    if adjust:
        ret = ret * 100000/ pop[st]
    
    return ret

%matplotlib inline

in_order("CT",value=["dv_int"]).plot()
```

```python
in_order("US").plot()
```

```python
in_order("CT",adjust=True,value=["pr_int"]).join(in_order("US", adjust=True,value=["pr_int"]),lsuffix="_CT", rsuffix="_US").plot()
```

```python
for_plot = in_order("CT",adjust=True,value=["pr_int"])\
.join(in_order("US", adjust=True,value=["pr_int"]),lsuffix="_CT", rsuffix="_US")
```
